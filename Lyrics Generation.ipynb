# load the pre-trained BERT from google drive

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

#load model
from transformers import BertForSequenceClassification, BertTokenizer

model = BertForSequenceClassification.from_pretrained('/content/drive/MyDrive/PLP project/models/bert_mode_lyrics_only')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
# 定义函数进行分类预测
def classify_song_style(song_lyrics):
    inputs = tokenizer(song_lyrics, return_tensors="pt", truncation=True, padding=True, max_length=9999)
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        predicted_class_idx = torch.argmax(logits, dim=-1).item()

    # 返回预测的风格标签
    return "Pop" if predicted_class_idx == 0 else "Hip Hop"
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# 加载GPT-2模型和tokenizer
gpt2_model = GPT2LMHeadModel.from_pretrained("gpt2")
gpt2_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# 定义生成歌词的函数
def generate_lyrics(style, artist):
    prompt = f"Generate a song in the style of {artist} about {style}."

    inputs = gpt2_tokenizer.encode(prompt, return_tensors="pt")
    outputs = gpt2_model.generate(inputs, max_length=200, num_return_sequences=1, no_repeat_ngram_size=2, top_k=50)
    generated_lyrics = gpt2_tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_lyrics
```

```python
import streamlit as st

# 用户输入歌手和风格
st.title("歌词生成器")
song_style = st.selectbox("选择风格", ["love", "struggle", "party", "heartbreak"])
artist = st.selectbox("选择歌手", ["Eminem", "Adele", "Drake", "Beyoncé"])

# 输入歌词（可选）
song_lyrics_input = st.text_area("输入歌词（可选）", "")

# 预测歌手风格（如果提供了歌词）
if song_lyrics_input:
    predicted_class_idx = classify_song_style(song_lyrics_input)
    genre = ["pop","hip-hop"][predicted_class_idx]

# 根据选择的歌手和风格生成歌词
if st.button("生成歌词"):
    lyrics = generate_lyrics(song_style, artist)  # 使用GPT-2
    # lyrics = generate_lyrics_gpt3(song_style, artist)  # 使用GPT-3
    st.write(lyrics)
```
